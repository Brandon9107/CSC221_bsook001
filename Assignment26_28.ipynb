```python
#cell 9
def sieve_of_eratosthenes(limit):
    primes = [True] * (limit + 1)
    primes[0] = primes[1] = False

    p = 2
    while p * p <= limit:
        if primes[p] == True:
            for i in range(p * p, limit + 1, p):
                primes[i] = False
        p += 1

    prime_numbers = [i for i in range(2, limit + 1) if primes[i] == True]
    return prime_numbers

# Generating prime numbers from 2 to 1000
primes_2_to_1000 = sieve_of_eratosthenes(1000)
print(primes_2_to_1000)
```
